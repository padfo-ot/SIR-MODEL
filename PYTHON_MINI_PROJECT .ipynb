```python
#SIR model COVID-19
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
from scipy.integrate import odeint
import plotly.express as ps


def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta(t) * S * I / N
    dIdt = beta(t) * I * S/N -  gamma * I
    dRdt =  gamma * I
    
    return dSdt, dIdt, dRdt


N =int(input (" Enter the population ")) # approximate population of Chennai
D = 12 # It takes an average of two weeks to recover
gamma = 1/(D)

t1=int(input(" On which day was the city wide lockdown initiated? "))
t2=int(input(" On which day was the enchanced quarantine stage was enforced? "))
      
def R_0(t):
    if t<t1:
        return 4.0
    elif (t>t1 and t<t2):
        return 2.69
    else:
        return 1.2
        
def beta(t):
    return R_0(t) * gamma

S0, I0, R0 = N-1, 1, 0 # initial conditions: one exposed

t= np.linspace(0,200,100,dtype=int)
y0 = S0,I0, R0 
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T


time=np.linspace(0,200,100,dtype=int)
Suceptible=[]
Infected=[]
Recovered=[]


for i in t:
    Suceptible.append(S)
    Infected.append(I)
    Recovered.append(R)
    
suceptible= Suceptible[0]
infected=Infected[0]
recovered=Recovered[0]
    
a={"Day":time,"Suceptible":suceptible}
b={"Day":time,"Infected":infected}
c={"Day":time,"Recovered":recovered}



df1 = pd.DataFrame(a)
df2 = pd.DataFrame(b)
df3 = pd.DataFrame(c)


def plot_model(t,S,I,R):
    f,ax = plt.subplots(1,1,figsize=(10,4))
    ax.plot(t, S, 'b',linestyle = 'solid', alpha=0.8, linewidth=2, label='Susceptible')
    ax.plot(t, I, 'r',linestyle = 'solid', alpha=0.7, linewidth=2, label='Infected')
    ax.plot(t, R, 'g',linestyle = 'solid', alpha=0.7, linewidth=2, label='Recovered')
    ax.plot(t, S+I+R, 'c--', alpha=0.7, linewidth=2, label='Total')
    ax.set_xlabel('Time (days)')
    ax.set_ylabel('No.of people affected')
    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid(b=False, which='major', c='w', lw=2, ls='-')
    legend = ax.legend()
    legend.get_frame().set_alpha(0.5)
    for spine in ('top', 'right', 'bottom', 'left'):
       ax.spines[spine].set_visible(True)
    
    fig = ps.line(df1, x="Day", y="Suceptible")
    fig2 = ps.line(df2, x="Day", y="Infected")
    fig3 = ps.line(df3, x="Day", y="Recovered")
    
    plt.show()
    fig.show()
    fig2.show()
    fig3.show()
   
    
plot_model(t,S,I,R)
```
